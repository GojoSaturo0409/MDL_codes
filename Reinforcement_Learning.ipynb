## Part A

```python
# Defining Constant Values
Reward_Goal = 1
Penalty_Red = -1
Step_Cost = -0.04
Prob_Dirn_Act = 0.7
Prob_Dirn_Perp_Act = 0.15
Gamma = 0.95
Theta = 0.0001
A = ['up', 'down', 'left', 'right']
res = [[],[],[],[]]
count = 0
```

```python
# Defining next state for each state under the influence of different actions
def s_next(s, a):
    if s == 0:
        if (a == 'up'):
            return 3
        elif (a == 'right'):
            return 1
        else:
            return 0
    elif s == 1:
        if (a == 'left'):
            return 0
        elif (a == 'right'):
            return 2
        else:
            return 1
    elif s == 2:
        if (a == 'up'):
            return 5
        elif (a == 'left'):
            return 1
        else:
            return 2
    elif s == 3:
        if (a == 'up'):
            return 6
        elif (a == 'down'):
            return 0
        else:
            return 3
    elif s == 4:
        return 4
    elif s == 5:
        if (a == 'up'):
            return 8
        elif (a == 'down'):
            return 2
        else:
            return 5
    elif s == 6:
        if (a == 'up'):
            return 9
        elif (a == 'down'):
            return 3
        elif (a == 'right'):
            return 7
        else:
            return 6
    elif s == 7:
        if (a == 'up'):
            return 10
        elif (a == 'left'):
            return 6
        elif (a == 'right'):
            return 8
        else:
            return 7
    elif s == 8:
        if (a == 'up'):
            return 11
        elif (a == 'down'):
            return 5
        elif (a == 'left'):
            return 7
        else:
            return 8
    elif s == 9:
        if (a == 'right'):
            return 10
        elif (a == 'down'):
            return 6
        else:
            return 9
    elif s == 10:
        if (a == 'left'):
            return 9
        elif (a == 'right'):
            return 11
        elif (a == 'down'):
            return 7
        else:
            return 10
    elif s == 11:
        if (a == 'down'):
            return 8
        elif (a == 'left'):
            return 10
        else:
            return 11
        
```

```python
def Psa(next_s, s, a):
    Prob = 0
    # If next state is same as the current one
    if (next_s == s):
        if (a == 'up'):
            if (s == 9 or s == 10 or s == 11 or s == 1):
                Prob = Prob_Dirn_Act
        elif (a == 'down'):
            if (s == 0 or s == 1 or s == 2 or s == 7):
                Prob = Prob_Dirn_Act
        elif (a == 'left'):
            if (s % 3 == 0 or s == 5):
                Prob = Prob_Dirn_Act
        elif (a == 'right'):
            if (s % 3 == 2 or s == 3):
                Prob = Prob_Dirn_Act
    else:    # Probability of next state under the influence of different actions
        if (a == 'up'):
            if (next_s == s+3):
                Prob = Prob_Dirn_Act
            elif ((next_s == s+1) or (next_s == s-1)):
                Prob = Prob_Dirn_Perp_Act
            else:
                Prob = 0
        elif (a == 'down'):
            if (next_s == s-3):
                Prob = Prob_Dirn_Act
            elif ((next_s == s+1) or (next_s == s-1)):
                Prob = Prob_Dirn_Perp_Act
            else:
                Prob = 0
        elif (a == 'left'):
            if (next_s == s-1):
                Prob = Prob_Dirn_Act
            elif ((next_s == s+3) or (next_s == s-3)):
                Prob = Prob_Dirn_Perp_Act
            else:
                Prob = 0
        elif (a == 'right'):
            if (next_s == s+1):
                Prob = Prob_Dirn_Act
            elif ((next_s == s+3) or (next_s == s-3)):
                Prob = Prob_Dirn_Perp_Act
            else:
                Prob = 0
    return Prob
```

```python
def value_iteration(Theta, Gamma):
    V = [0] * 12
    V[4] = 0
    V[11] = Reward_Goal
    V[10] = Penalty_Red
    V_new = [0] * 12
    V_new[11] = Reward_Goal
    V_new[10] = Penalty_Red
    count = 0
    Directions = [0] * 12
    Directions[4] = 'none'
    Directions[11] = 'none'
    Directions[10] = 'none'

    while True:
        count = count + 1
        for s in range(12):

            if (s != 4 and s != 10 and s != 11):
                Utility = {'up': 0, 'down': 0, 'left': 0, 'right': 0}
                for a in A:
                    for action in A:
                        Next_State = s_next(s, action)
                        if (Next_State == s):
                            if (a == action):
                                Utility[a] += Prob_Dirn_Act*V[Next_State]
                            elif ((a == 'up' and action == 'down') or (a == 'left' and action == 'right') or (a == 'right' and action == 'left') or (a == 'down' and action == 'up')):
                                Utility[a] += 0
                            else:
                                Utility[a] += Prob_Dirn_Perp_Act*V[Next_State]
                        else:
                            Utility[a] += Psa(Next_State, s, a)*V[Next_State]

                key = max(Utility, key=Utility.get)
                Directions[s] = key
                V_new[s] = Step_Cost + Gamma*max(Utility.values())

        if all(abs(V_new[s] - V[s]) < Theta for s in range(12)):
            V = V_new
            break

        V = V_new.copy()
        res[0] = V_new[9:12]
        res[1] = V_new[6:9] 
        res[2] = V_new[3:6]
        res[3] = V_new[0:3]

        print(count)
        for ww in range(4):
            print(res[ww])
            


    return V
```

```python
qq = [[],[],[],[]]

r1 = value_iteration(Theta,Gamma)
print("Final Iteration:")
qq[0] = r1[9:12]
qq[1] = r1[6:9]
qq[2] = r1[3:6]
qq[3] = r1[0:3]
for ww in range(4):
    print(qq[ww])
```
